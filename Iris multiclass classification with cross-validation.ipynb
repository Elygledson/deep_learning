```
import pandas as pd
import numpy as np
import torch.nn as nn
from skorch import NeuralNetClassifier
import torch
from sklearn.model_selection import cross_val_score
torch.__version__
```

## Step 2: Database

```
np.random.seed(123)
torch.manual_seed(123)
```

```
dataframe = pd.read_csv('iris.csv')
dataframe.drop(columns='Id', inplace=True)
```

```
dataframe.head()
```

```
features = dataframe.iloc[:, 0:4].values
classes = dataframe.iloc[:, 4].values
```

```
classes
```

```
np.unique(classes)
```

```
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
classe = encoder.fit_transform(classes)
```

```
np.unique(classe)
```

```
features = features.astype('float32')
classe = classe.astype('int64')
```

## Step 3: Building model

```
class classificador_torch(nn.Module):
    def __init__(self):
        super().__init__()
        self.dense0 = nn.Linear(4, 16)
        self.activation0 = nn.ReLU()
        self.dense1 = nn.Linear(16, 16)
        self.activation1 = nn.ReLU()
        self.dense2 = nn.Linear(16, 3)

    def forward(self, X):
        X = self.dense0(X)
        X = self.activation0(X)
        X = self.dense1(X)
        X = self.activation1(X)
        X = self.dense2(X)
        return X
```

```
classifier_sklearn = NeuralNetClassifier(module = classificador_torch,
                                            criterion = torch.nn.CrossEntropyLoss,
                                            optimizer = torch.optim.Adam,
                                            max_epochs = 1000,
                                            batch_size = 10,
                                            train_split = False)
```

## Step 4: Cross Validation

```
results = cross_val_score(classifier_sklearn, features, classe, cv = 5,
                             scoring = 'accuracy')
```
